# **Q-Learning**

### 1. Introduction

So far, we have considered _model-based_ RL where we assign mathematical structure to the environment. In MDPs, the environment was modelled as a Markov chain with rewards. Moreover, a probability transition matrix modelled the environment. However, other methods in RL can be described as _model-free_ and do not have an internal model of the environment. Model-free methods learn directly through data and make decisions based on past experiences without a complete world model. Compared to model-based methods, model-free methods are computationally less demanding, but require more training data. In this notebook, we will dicuss one of the most influential model-free algorithms in RL: **Q-learning**.

# **Introduction to Reinforcement Learning**

### 1. Introduction

Reinforcement Learning (RL) is a form of machine learning in which an agent learns how to act by interacting with an environment. Interactions include making decisions (actions), receiving feedback (rewards) and continually improving its behaviour to maximise total reward over time. Many models in RL have been inspired by studies of human and animal behaviour in biology and psychology. [...]. Essentially all RL algorithms follow the loop below,

- The agent observes the current state $s$ of the environment.
- After observing the state $s_t$, the agent chooses an action $a$.
- The environment responds by modifying the state of the algorithm to $s_{t+1}$ while also providing a reward $r$ to the agent.
- The agent updates its internal representation based on the reward $r$ to amend its knowledge and strategy.

Mathematically, we have,

$$
s_t \rightarrow a_t \rightarrow r_t, \; s_{t+1}
\rightarrow a_{t+1} \rightarrow r_{t+1}, \; s_{t+2}
\rightarrow \cdots
$$

where $s_t$ is our initial state of the environment.

### 2. Multi-Armed Bandit

A one-armed bandit machine are the classic casino slot machines popularised in media. They feature three rolling images which are rotated by a single lever.   

There are three main componenets to a reinforcement learning task:

- Action $a$ - Pulling the lever on a bandit machine.
- Expected reward $r$.
- The exploitation-exploration trade-off - Should we continue using the current bandit machine or switch to another?

Regret, $\rho$, is the difference between the average payoff using the chosen RL algorithm and the maximum payoff achievable from knowing all information prior.

$\epsilon$-Greedy: The $\epsilon$-Greedy strategy aims to find a balance between exploration and exploitation. $\epsilon$ controls the exploitation-exploration trade-off with larger $\epsilon$ favouring more exploration. In the context of the multi-armed bandit problem, it is the probability of the agent selecting a random bandit machine that has not shown the high average payoff over all previous rounds on a given iteration such that $0 < \epsilon < 1$.

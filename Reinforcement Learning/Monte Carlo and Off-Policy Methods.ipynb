# **Monte Carlo Methods**

### 1. Introduction

So far, we have considered _model-based_ RL where we assign mathematical structure to the environment. In MDPs, the environment was modelled as a Markov chain with rewards. A probability transition matrix modelled the environment since $p(s', r \vert s, a)$ for all state-action pairs $(s, a)$ was known. However, other methods in RL can be described as _model-free_ and do not have an internal model of the environment. Model-free methods learn directly through data and make decisions based on past experiences without a complete world model. Compared to model-based methods, model-free methods are computationally less demanding, but require more training data. In this notebook, we will discuss a family of the most basic model-free algorithm in RL: **Monte Carlo "MC" Methods**.

MC methods use averages over partial trajectories to approximate estimate values. These methods do NOT estimate $p(s', r \vert s, a)$. In other words, they do not attempt to built a model from sampling the trajectory after running some policy. Recall that in model-based RL we have, 

$$
Q(s,a) = \sum_{s' \in \mathcal{S}} \sum_{r \in \mathcal{R}} p(s', r \mid s, a) \underbrace{\left[r + \gamma \max_{a' \in \mathcal{A}(s')} Q(s', a') \right]}_{V(s')}
$$

which gives the state-action function $q(s, a)$ for the optimal policy $\Pi*$. Since MC method assume $p(s', r \mid s, a)$ is unknown, $Q(s,a)$ must be estimated directly through observations. However, it should be noted that $Q(s,a)$ is not directly estimated. Instead, $V(s)$ is directly estimated because it is possible to express $q_{\Pi}$ estimation as $v_{\Pi}$ estimation. In doing so, we consider a Markov reward process (MRP) rather than a MDP by _"absorbing"_ actions into the environment. This simplifies our analysis. Therefore, the main idea of MC methods is to estimate $v_{\Pi}$ for some policy $\Pi$ without constructing a MDP model and then using a form of policy improvement (GPI) to converge to the optimal $\Pi^*$.

### 2. Basic Monte Carlo Methods

As mentioned before, the main idea of Monte Carlo (MC) methods is to estimate $v_{\Pi}$(s) by averaging observed returns from sampled episodes generated under policy $\Pi$ in a Markov Reward Process (MRP) induced by $\Pi$. Our method remains model-free because we do not attempt to construct the transitions dynamics $P^{\Pi}(s', r \vert s)$. That is, MC methods only use information from the observations and not the internal model of the environment. MC methods only see $M$ trajectory samples of an MRP (we consider this the data):

$$
s_0^{(m)}, r_1^{(m)}, s_1^{(m)}, r_2^{(m)}, ..., s^{m}_{T_m} \quad m = 1, ..., M
$$

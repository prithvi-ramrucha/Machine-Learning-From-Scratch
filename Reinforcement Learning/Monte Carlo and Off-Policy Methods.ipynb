# **Monte Carlo Methods**

### 1. Introduction

So far, we have considered _model-based_ RL where we assign mathematical structure to the environment. In MDPs, the environment was modelled as a Markov chain with rewards. A probability transition matrix modelled the environment since $p(s', r \vert s, a)$ for all state-action pairs $(s, a)$ was known. However, other methods in RL can be described as _model-free_ and do not have an internal model of the environment. Model-free methods learn directly through data and make decisions based on past experiences without a complete world model. Compared to model-based methods, model-free methods are computationally less demanding, but require more training data. In this notebook, we will discuss a family of the most basic model-free algorithm in RL: **Monte Carlo "MC" Methods**.

MC methods use averages over partial trajectories to approximate estimate values. These methods do NOT estimate $p(s', r \vert s, a)$. In other words, they do not attempt to built a model from sampling the trajectory after running some policy. Recall that in model-based RL we have, 

$$
Q(s,a) = \sum_{s' \in \mathcal{S}} \sum_{r \in \mathcal{R}} p(s', r \mid s, a) \underbrace{\left[r + \gamma \max_{a' \in \mathcal{A}(s')} Q(s', a') \right]}_{V(s')}
$$

which gives the state-action function $q(s, a)$ for the optimal policy $\Pi^*$. Since MC method assume $p(s', r \mid s, a)$ is unknown, $Q(s,a)$ must be estimated directly through observations. This is the true in Monte Carlo control methods (the most general form of MC in RL). However, in Monte Carlo evaluation or prediction $Q(s,a)$ is not directly estimated. Instead, $V(s)$ is directly estimated because the policy $\Pi$ is kept fixed. In doing so, we consider a Markov reward process (MRP) rather than a MDP by _"absorbing"_ actions into the environment. This simplifies our analysis. Therefore, the main idea of MC evaluation is to estimate $v_{\Pi}$ for some policy $\Pi$ without constructing a MDP model and then using a form of policy improvement (GPI) to converge to the optimal $\Pi^*$. MC control, on the other hand, seeks to solve the full reinforcement learning problem directly. Rather than estimating only the value of a fixed policy, MC control estimates the action-value function $q_{\Pi}(s,a)$ from sampled experience and uses these estimates to improve the policy. By repeatedly evaluating action values and selecting actions with higher estimated returns, the policy gradually becomes more optimal. In the limit, MC control aims to recover both the optimal action-value function $Q(s,a)$ and the corresponding optimal policy $\Pi^*$ without ever requiring knowledge of the transition probabilities $p(s', r \mid s, a)$.

### 2. Basic Monte Carlo Evaluation

As mentioned before, the main idea of Monte Carlo (MC) methods is to estimate $v_{\Pi}$(s) by averaging observed returns from sampled episodes generated under policy $\Pi$ in a Markov Reward Process (MRP) induced by $\Pi$. Our method remains model-free because we do not attempt to construct the transitions dynamics $P^{\Pi}(s', r \vert s)$. That is, MC methods only use information from the observations and not the internal model of the environment. MC methods only see $M$ trajectory samples of an MRP (we consider this the data):

$$
s_0^{(m)}, r_1^{(m)}, s_1^{(m)}, r_2^{(m)}, ..., s^{m}_{T_m} \quad m = 1, ..., M
$$

where the superscript $m$ denotes the $m$-th trajectory and $T_m$ is the number of epsiodes. Given that by definition, $v_{\pi}$ is the expected discounted return starting from the state $s$,

$$
v_{\pi}(s) = E_{\Pi} [G_t \vert S_t = s] \approx \frac{1}{C(s)} \sum_{m=1}^M \sum_{\tau=0}^{T_m - 1} \mathcal{I}(s_\tau^m = s)g_{\tau}^m
$$

where $g^{(m)}_{\tau} = r^{(m)}_{\tau + 1} + \gamma r^{(m)}_{\tau + 2} + + \gamma^2 r^{(m)}_{\tau + 3} + ... + \gamma^{T_m - 1 - \tau}  r^{(m)}_{T_m}$ is the discounted return given by $s$ and future states along the trajectory. The idea is to sum over all trajectories and their episodes to pick out the future returns of a specificied state $s$. Note that $\mathcal{I}(s_\tau^m = s)$ is an indicator function that is only non-zero and equal to `1` when $s_\tau^m = s$. $C(s)$ is the counting function and gives the number of times the state $s$ is present across all trajectories. The explicit form for $C(s)$ is,

$$
C(s) = \sum_{m=1}^M \sum_{\tau=0}^{T_m - 1} \mathcal{I}(s_\tau^m = s)
$$

We can imagine the indicator function $\mathcal{I}$ as a switch which returns `1` when its found an episode with state $s$ and `0` otherwise. This way of estimating $E_{\Pi} [G_t \vert S_t = s]$ is known as **_every-visit_** MC because we count all visits to $s$ across the trajectories. One can show that in the special case where only rewards associated with terminal states are non-zero such that $r^{(m)}_{k} = 0$ for all $k \neq T_m$ and $\gamma = 1$ (no discounting), the estimator reduces to,

$$
\boxed{
v_{\pi}(s) \approx \frac{\sum_{m=1}^{M} r_m \, N_m(s)}{\sum_{m=1}^{M} N_m(s)}
}
$$

where $N_m(s)$ is the number of times the state $s$ is occurs in the $m$-th trajectory and the counting function $C(s)$ is the sum of $N_m(s)$ across all trajectories such that $C(s) = \sum_m N_m(s)$. In addition to the every-visit estimator, there exists a _first-visit_ estimator which only considers the first visit to the state $s$,

$$
v_{\pi}(s) = E_{\Pi} [G_t \vert S_t = s] \approx \frac{1}{C(s)} \sum_{m=1}^M \sum_{\tau=0}^{T_m - 1} \mathcal{I}(s_\tau^m = s\;\land\;s_k^m \neq s,\ \forall k < \tau)g_{\tau}^m
$$

In other words, once the indicator function $\mathcal{I}$ has found the virst occurance of $s$ in the a given trajectory, it returns `1` and then `0` for all other states (including more occurrences of $s$). Let us say that 


$$
v_{\pi}(s) = E_{\Pi} [G_t \vert S_t = s] \approx \frac{1}{M} \sum_{m=1}^M g_{k(m)}^m
$$

# **Introduction to Principal Component Analysis (PCA)**

### 1. Mathematical Overview

PCA is the most well-known algorithm for dimensionality reduction. It is a unsupervised learning method and transforms a higher dimensional dataset into a lower one with the goal of perserving as much information as possible. It is also useful in exploratory data analysis (EDA), because it allows for the visualisation of high dimensional data. In PCA we want to map the set of $N$ data points where each data point $X_j = (x_{j1}, x_{j2}, x_{j3}, ..., x_{ND})$ exists in the space $\mathbb{R}^{D}$ with $D$ number of dimensions to the set of points where $\mathcal{X_j} = (\mathcal{X_{j1}}, \mathcal{X_{j2}}, \mathcal{X_{j3}}, ..., \mathcal{X}_{NK})$ in $\mathbb{R}^{K}$ where $K<D$. We can significately reduce the storage size of a dataset without major information loss using PCA. The transformation from $X_{jk}$ to $\mathcal{X}_{jk}$ is given by,

$$\mathcal{X}_{jk} = \mathbf{P}_k^{T} X_{jk}$$

$\mathbf{P}_k$ are our principal components which are linear combinations of our original features $x_{j1}$, $x_{j2}$, $x_{j3}$, etc. To determine our principle components, we must solve the following eigenvalue equation,

$$\Sigma \mathbf{P}_k = \lambda_k \mathbf{P}_k$$

In which, $\Sigma$ represents the covariance matrix, $\mathbf{P}_k$ is the $k$-th principal component and $\lambda_k$ is its corresponding eigenvalue. The elements of the covariance matrix are given by, 

$$ \Sigma_{\mu \nu} = \frac{1}{N} \sum^{N}_{i=j} (x_{j\mu} - \langle x_{\mu} \rangle)^T (x_{j\nu} - \langle x_{\nu} \rangle)$$

$\mu$ and $\nu$ are indices that index features of the original dataset. Therefore, $\Sigma$ is a square matrix with the dimensions $[D \times D]$. Wheb solving the eigenequation, one will find that there are $D$ number of eigenvalues for the same number of principal components such that,

$$\lambda_1 > \lambda_2 > ... > \lambda_k > ... > \lambda_D$$

In PCA, we choose how may dimensions we want the reduced dataset to have. Hence, if we want to reduce the dataset down to $K$ dimesions, we selected the $K$ largest eigenvalues $\lambda$ and their respective principal components (eigenvectors). This, in essence, is the PCA algorithm. However, the mathematical steps required to produce these results requires many moving parts. In other words, deriving PCA from first principles is mathematically rich. From a high-level view, PCA makes the assumption that information loss is minimised when the variance between the projected points from the higher dimensional space (our original dataset) to the lower dimensional space (transformed dataset) is maximised. A spread of values implies that there has been minimial compression and information loss. A deriviation of PCA from first principles in provided in the final section of this notebook.


### 2. High-Level Intuition 

PCA reduces the dimensionality of data by finding the principle components which explain the most variance. But what exactly does this mean? Say that we have the points $\mathbf{x}_i = (x_{1i}, x_{2i})$. When we visualise our two dimensional feature space, intuitively we can see that a straight line fits the data reasonably well. The direction of this line is assigned to a principle component $\mathbf{P}_1$. We say that the line drawn in the direction of $\mathbf{P}_1$ minimises the sum of squared perpendicular distances from each point to it. It can be shown that this is equivalent to maximising the variance.

### 4. PCA From First Principles

The most intuitive method of deriving PCA from first principles is to consider the maximisation of variance when projecting data. Once again, let us consider our elementary example where we want to project 2D data samples $\mathbf{x_i} = (x_{i1}, x_{i2})$ onto a single dimension. This single dimension is defined by a line whose directon is given by the unit vector $\mathbf{u}$. We therefore seek the vector $\mathbf{u}$ that maximises the variance of the projected samples onto this line, such that the resulting one-dimensional representation retains as much information from the original dataset as possible. The projection of a single sample $\mathbf{x_i}$ on $\mathbf{u}$ is given by $z_i = \mathbf{u}^T \mathbf{x_i}$ such that the mean over all projections is,

$$
\mathbf{E}[z] = \frac{1}{N} \sum^{N}_{i=1} \mathbf{u}^T \mathbf{x_i} = \mathbf{u}^T \left(\frac{1}{N} \sum^{N}_{i=1} \mathbf{x_i} \right) = \mathbf{u}^T \mathbf{\mu}
$$

where $\mathbf{\mu}$ represents the vector mean over all samples $\mathbf{x_i}$. By definition, the variance of the projections $z_i$ is,

$$
\text{Var}(z) = \mathbf{E}[\left( z - \mathbf{E}[z] \right)^2] = \frac{1}{N} \sum^{N}_{i=1} (\mathbf{u}^T \mathbf{x_i} - \mathbf{u}^T \mathbf{\mu})^2
$$

Using vector properties and the definition for the closed form of the covariance matrix, we can re-write $\text{Var}(z)$ as follows,

$$
\text{Var}(z) = \frac{1}{N} \sum^{N}_{i=1} \left( \mathbf{u}^T(\mathbf{x_i} - \mathbf{\mu}) \right)^2
$$

$$
 = \frac{1}{N} \sum^{N}_{i=1} \left( \mathbf{u}^T(\mathbf{x_i} - \mathbf{\mu}) \right)^2
$$

$$
 = \frac{1}{N} \sum^{N}_{i=1} \mathbf{u}^T(\mathbf{x_i} - \mathbf{\mu}) (\mathbf{x_i} - \mathbf{\mu})^T \mathbf{u}
$$

$$
 = \mathbf{u}^T \left( \frac{1}{N} \sum^{N}_{i=1} (\mathbf{x_i} - \mathbf{\mu}) (\mathbf{x_i} - \mathbf{\mu})^T \right) \mathbf{u}
$$

$$
 = \mathbf{u}^T S \mathbf{u}
$$

We have used the fact that the tranpose of a scalar is itself so $\mathbf{u}^T(\mathbf{x_i} - \mathbf{\mu}) = (\mathbf{u}^T(\mathbf{x_i} - \mathbf{\mu}))^T = (\mathbf{x_i} - \mathbf{\mu})^T\mathbf{u}$ which employes the identity $(\mathbf{a}\mathbf{b})^T = \mathbf{b}^T\mathbf{a}^T$. $S$ denotes the closed form of the covariance matrix which is defined as,

$$
S = \frac{1}{N} \sum^{N}_{i=1} (\mathbf{x_i} - \mathbf{\mu}) (\mathbf{x_i} - \mathbf{\mu})^T 
$$

$$
= \begin{bmatrix}
\frac{1}{N}\sum_{i=1}^N (x_{i1}-\mu_1)^2 & \frac{1}{N}\sum_{i=1}^N (x_{i1}-\mu_1)(x_{i2}-\mu_2) & \cdots & \frac{1}{N}\sum_{i=1}^N (x_{i1}-\mu_1)(x_{id}-\mu_D) \\
\frac{1}{N}\sum_{i=1}^N (x_{i2}-\mu_2)(x_{i1}-\mu_1) & \frac{1}{N}\sum_{i=1}^N (x_{i2}-\mu_2)^2 & \cdots & \frac{1}{N}\sum_{i=1}^N (x_{i2}-\mu_2)(x_{id}-\mu_D) \\
\vdots & \vdots & \ddots & \vdots \\
\frac{1}{N}\sum_{i=1}^N (x_{iD}-\mu_D)(x_{i1}-\mu_1) & \frac{1}{N}\sum_{i=1}^N (x_{id}-\mu_D)(x_{i2}-\mu_2) & \cdots & \frac{1}{N}\sum_{i=1}^N (x_{id}-\mu_D)^2
\end{bmatrix}
$$

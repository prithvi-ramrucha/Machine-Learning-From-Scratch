# Hierarchical Clustering

### Theory 

Hierarchical clustering is a unsupervised algorithm that assigns labels $\hat{y}_k$ to unlabelled data $\mathbf{x_i}$. It builds groups step-by-step, either by merging smaller clusters or splitting larger ones. The algorithm often produces intuitive groupings but is myopic in the sense that it is greedy and does not consider global structure. A bottom-up approach (agglomerative), where each individual data points starts off as a cluster, is the most common method for hierarchical clustering. However, a divisive (top-down) approach in which all points are initially assigned to a single cluster is also used. Although, this method is more computationally expensive. For this notebook, we will focus on the former approach. The steps for agglomerative hierarchical clustering are as follows,

1. Initialise the clusters that such each data point is assigned to its own cluster.
2. We merge the two clusters with the smallest distance between them over all possible pairwise distances between the cluster centroids.
3. The previous step is repeated until only a single cluster is left. 



### Limitations

Bottom-up hierarchical clustering is one of the most popular clustering algorithms because it often produces good results and its highly interpretable. Despite this, it does have significant short-comings. As mentioned previously, the algorithm for hierarchical clustering is myopic. This means that it is greedy and makes only locally optimal decisions at each step. The result may be misleading cluster structures which due not reflect the true cluster structure of the data. Moreover, hierarchical clustering is computationally expensive with $O(N^3)$ time complexity and $O(N^2)$ in memory. It also suffers from the curse of dimensionality where distance metrics lose meaning as the dimensionality of the data increases. For these reasons, hierarchical clustering is best suited for small, clean and low-dimensional data sets for which interpretability is a priority.

**Customer Segmentation Example**: A clear example is customer segmentation. Customers may naturally group by one factor (e.g. purchasing behaviour), but early merges based on another factor (e.g. spending volume) can mix different behavioural types into the same cluster. Once this happens, it becomes impossible to later separate them into more meaningful groups. This shows how early decisions can distort the final clustering outcome.
